This function is to plot the confusion matrix.


```python
def plot_confusion_matrix(y,y_predict):
    "this function plots the confusion matrix"
    from sklearn.metrics import confusion_matrix

    cm = confusion_matrix(y, y_predict)
    ax= plt.subplot()
    sns.heatmap(cm, annot=True, ax = ax); #annot=True to annotate cells
    ax.set_xlabel('Predicted labels')
    ax.set_ylabel('True labels')
    ax.set_title('Confusion Matrix'); 
    ax.xaxis.set_ticklabels(['did not land', 'land']); ax.yaxis.set_ticklabels(['did not land', 'landed']) 
    plt.show() 
```

## TASK  12


Find the method performs best:


```python


# Define models and their parameter grids
models_param_grids = {
    'Decision Tree': {
        'model': DecisionTreeClassifier(),
        'params': {
            'criterion': ['gini', 'entropy'],
            'max_depth': [None, 10, 20, 30]
        }
    },
    'KNN': {
        'model': KNeighborsClassifier(),
        'params': {
            'n_neighbors': [3, 5, 7],
            'weights': ['uniform', 'distance']
        }
    },
    'Logistic Regression': {
        'model': LogisticRegression(),
        'params': {
            'penalty': ['l1', 'l2'],
            'C': [0.1, 1, 10],
            'solver': ['liblinear']
        }
    },
    'svc':{
         'model' :SVC(),
         'params' : {
             'kernel':('linear', 'rbf','poly','rbf', 'sigmoid'),
              'C': np.logspace(-3, 3, 5),
              'gamma':np.logspace(-3, 3, 5)
            
           }
    }
}
# Initialize a dictionary to store results
results = {}

# Perform GridSearchCV for each model
for model_name, mp in models_param_grids.items():
    grid_search = GridSearchCV(mp['model'], mp['params'], cv=10, scoring='accuracy')
    grid_search.fit(X_train, Y_train)
    results[model_name] = {
        'best_params': grid_search.best_params_,
        'validation_score': grid_search.best_score_,
        'best_estimator': grid_search.best_estimator_
    }

# Find the best performing model
best_model_name = max(results, key=lambda k: results[k]['validation_score'])
best_model = results[best_model_name]

# Print the results
print("Best Performing Model:")
print(f"Model: {best_model_name}")
print(f"Best Parameters: {best_model['best_params']}")
print(f"Best Validation Accuracy: {best_model['validation_score']}")

# Evaluate the best model on the test data
test_accuracy = best_model['best_estimator'].score(X_test, Y_test)
print(f"Test Accuracy for {best_model_name}: {test_accuracy}")
```
